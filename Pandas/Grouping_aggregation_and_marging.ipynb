```python
import pandas as pd
```

```python
str= r'C:\Users\KIIT\OneDrive\Desktop\code 1\Data Analysis\BD-Dataset.csv'
date_bd= pd.read_csv(str)
date_bd
```

```python
date_bd['date']=pd.to_datetime(date_bd.Date)
date_bd
```

```python
date_bd['year']=pd.DatetimeIndex(date_bd.date).year
date_bd['month']=pd.DatetimeIndex(date_bd.date).month
date_bd['day']=pd.DatetimeIndex(date_bd.date).day
date_bd['weekday']=pd.DatetimeIndex(date_bd.date).weekday
date_bd
```

```python
date_bd.drop(columns=['Date'],inplace=True)
date_bd
```

```python
group_date= date_bd.groupby('month')
type(group_date)
```

```python
group_date[['Daily new confirmed cases','Daily new deaths','Daily new recovered']].sum()
```

```python
weekday_grouped= date_bd.groupby('weekday')[['Daily new confirmed cases','Daily new deaths','Daily new recovered']].mean()
weekday_grouped
```

```python
location_df=pd.read_csv('countries.csv')
location_df
```

```python
location_df[location_df.location=='Bangladesh']
```

```python
date_bd['location']="Bangladesh"
date_bd
```

```python
merged_df= date_bd.merge(location_df,on="location")
merged_df
```

```python
merged_df['deaths_per_million']=merged_df['Total deaths'] * 1e6/ merged_df.population
merged_df['Tests_per_million']=merged_df['Total Tests'] * 1e6/ merged_df.population
merged_df
```

```python
result_df= merged_df[[
                    'date',
                    'Daily New Tests',
                    'Total Tests',
                    'Daily new deaths',
                    'Total deaths',
                    'deaths_per_million',
                    'Tests_per_million',
                    'Daily new recovered',
                    'Total recovered']]
result_df
```

```python
result_df.to_csv('results.csv',index=False)
```
